### Conteúdo

```python
df = pd.read_csv("../data/transacoes.csv", sep=";")
df.head()
```

Valores maior que 50

```python
filtro_pontos = df["QtdePontos"] >= 50
df[filtro_pontos]
```

Valores entre 50 e 100

```python
filtro_pontos = (df["QtdePontos"] >= 50) & (df["QtdePontos"] < 100)
df[filtro_pontos]
```

Valores iguais a 1 ou maior que 100

```python
filtro_or = (df["QtdePontos"] == 1) | (df["QtdePontos"] == 100)
df[filtro_or]
```

Pontos entre 0 e 50 e ano minimo 2025

```python
filtro_data = (df["QtdePontos"] > 0) & (df["QtdePontos"] <= 50) & (df["DtCriacao"] >= "2025-01-01")
df[filtro_data]
```
