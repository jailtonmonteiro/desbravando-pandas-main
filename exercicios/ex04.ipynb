```python
import pandas as pd
```

```python
clientes = pd.read_csv("../data/clientes.csv", sep=";")
transacoes = pd.read_csv("../data/transacoes.csv", sep=";")
```

```python
# Quantos clientes tem vínculo com a Twitch?
vinculo_Twitch = clientes["flTwitch"].sum()
vinculo_Twitch
```

```python
# Quantos clientes tem um saldo de pontos maior que 1000?
saldo_1000 = (clientes["qtdePontos"] > 1000).sum()
saldo_1000
```

```python
# Quantas transações ocorreram no dia 2025-02-01?
transacoes["DtCriacao"] = pd.to_datetime(transacoes["DtCriacao"]).dt.normalize()
ocorrencias = (transacoes["DtCriacao"] == "2025-02-01").sum()
ocorrencias
```
