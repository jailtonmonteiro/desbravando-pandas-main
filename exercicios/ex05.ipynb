```python
import pandas as pd
import numpy as np
```

```python
clientes = pd.read_csv("../data/clientes.csv", sep=";")
```

```python
# Crie uma coluna nova “twitch_points” que e resultado da multiplicação do saldo de pontos e a marcação da twitch
clientes["twitch_points"] = clientes["flTwitch"] * clientes["qtdePontos"]
clientes.head()
```

```python
# Aplique o log na coluna de saldo de pontos, criando uma coluna nova
clientes["logPontos"] = np.log(clientes["qtdePontos"])
clientes.head()
```

```python
# Crie uma coluna que sinalize se a pessoa tem vínculo com alguma (qualquer uma) plataforma de rede social.
clientes["social"] = clientes["flEmail"] | clientes["flTwitch"] | clientes["flYouTube"] | clientes["flBlueSky"] | clientes["flInstagram"]
clientes[["flEmail", "flTwitch", "flYouTube", "flBlueSky", "flInstagram", "social"]].head()
```

```python
# Qual é o id de cliente que tem maior saldo de pontos? E o menor?
clientes.sort_values(by="qtdePontos", ascending=False).head(1)
```

```python
clientes.sort_values(by="qtdePontos", ascending=True).head(1)
```

```python
# Selecione a primeira transação diária de cada cliente.

transacoes = pd.read_csv("../data/transacoes.csv", sep=";")

transacoes = transacoes.sort_values("DtCriacao")

transacoes["data"] = pd.to_datetime(transacoes["DtCriacao"]).dt.date
transacoes.head()

transacoes.drop_duplicates(keep="first", subset=["IdCliente", "data"])
```

```python
first = transacoes.drop_duplicates(keep="first", subset=["IdCliente", "data"])
last = transacoes.drop_duplicates(keep="last", subset=["IdCliente", "data"])

pd.concat([first, last])
```
